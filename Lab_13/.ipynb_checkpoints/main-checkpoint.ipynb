```python
#structure of a github request is in the link (seach files that end with .scaxxxx in this path+repo) ==> code?q=.scavengerhunt+in:path+repo:inronhack-datalabs/scavenger
# ?q = query
# + means another parameter
# inside a parameter you can put an "add"
# +in:path+repo: 
#the result is not only the name of the files but also different informations so the next code is to select the exact file element
response = requests.get('https://api.github.com/search/code?q=.scavengerhunt+in:path+repo:ironhack-datalabs/scavenger')
search = response.json()
search
```

```python
#the file name is in the header "items" so we select it
#we sort them
repo_files_names=[]
for i in search["items"]:
    repo_files_names.append(i["name"])
repo_files_names.sort()
print(repo_files_names)
```

```python
#we get the file and append its URL
repo_files_urls = {item["name"]: item["url"] for item in search["items"]}
repo_files_urls
```

```python
#Gets us the files
repo_files = [requests.get(repo_files_urls[repo_filename]) for repo_filename in repo_files_names]
repo_files
```

```python
#gets us the content foe each file, which is in an encoded format
repo_files_content = [file.json()["content"] for file in repo_files]
repo_files_content
```

```python
#this library allows us to decode a message

import base64
message = [base64.b64decode(b64message).decode('ascii') for b64message in repo_files_content]
message
```
