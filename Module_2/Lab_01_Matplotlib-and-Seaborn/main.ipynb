#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
fig, ax = plt.subplots(1,2,figsize = (12,10))
ax[0].set_title('Y function', fontsize=18)
ax[0].plot(x,y, label='y=x*2')
ax[1].set_title('Z function', fontsize=18)
ax[1].plot(x,z, label='y=x**2')
ax.subplots_adjust(wspace=1) # adjust space between 2 plots
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
import math
```

```python
x = np.arange(0,100)
y = x**2
z = math.e**x
```

```python
fig, ax = plt.subplots(1,2,figsize = (6,3))

ax1 = ax.flatten()[0]
ax1.plot(x,y, label='y=x*2')
ax1.plot(x,z, label='y=x**2')
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2 = ax.flatten()[1]
ax2.plot(x,y, label='y=x*2')
ax2.plot(x,z, label='y=x**2')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel("x")
ax2.set_ylabel("y")
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, ax = plt.subplots(1,2,figsize = (7,5))


ax[0].set_title('Y function', fontsize=18)
ax[0].plot(x,y, label='y=x*2')


ax[1].set_title('Z function', fontsize=18)
ax[1].plot(x,z, label='y=x**2')
ax[1].set_xscale('log')
ax[1].set_yscale('log')


ax[0].legend()
ax[1].legend()
```
