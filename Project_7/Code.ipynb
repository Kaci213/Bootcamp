```python
corr = df.corr()

def create_sns_heatmap(data):
    #need to import matplotlib.pyplot as plt
    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(data, dtype=bool))

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)

    sns.heatmap(data,mask=mask, cmap=cmap, annot=False,vmin=-0.5, vmax=0.5)
    return f,ax

create_sns_heatmap(corr)
```

```python
def drop_cor(bankrupcy, trsh):
    # Create correlation matrix
    corr_matrix = bankrupcy.corr().abs() # absolute value of correlation
    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),
                                k=1).astype(bool))
    # Find features with correlation greater than 0.90
    to_drop = [column for column in upper.columns if any(upper[column] > trsh)]
    # Drop features
    bankrupcy.drop(to_drop, axis=1, inplace=True)

drop_cor(df, 0.9)
df
```

# Modelling

```python
# Import the necessary libraries
from sklearn.model_selection import train_test_split


# Load the dataset
df = pd.read_csv('clean_bankrupt.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('bankrupt', axis=1), df['bankrupt'], test_size=0.2)
```

```python
# Balanced data on train set
y_train.value_counts()*100/len(y_train)
```

```python
from imblearn.over_sampling import SMOTE 
smote = SMOTE()
```

```python
X_sm, y_sm = smote.fit_resample(X_train, y_train)
y_sm.value_counts()
```

```python
y_sm.shape
```

```python
# scaling data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_sm)
X_train_mod = scaler.transform(X_sm)
X_test_mod  = scaler.transform(X_test)
```

```python
from sklearn.svm import LinearSVC, SVC, NuSVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
```

## LinearSVC

```python
lsvc = LinearSVC()
```

```python
lsvc.fit(X_train_mod, y_sm)
score = lsvc.score(X_train_mod, y_sm)
print("Score: ", score)
```

```python
y_train_pred = lsvc.predict(X_train_mod)
```

```python
cm_train = confusion_matrix(y_sm, y_train_pred)
print(cm_train)
```

```python
y_test_pred = lsvc.predict(X_test_mod)
```

```python
#cm_test = confusion_matrix(y_test, y_test_pred)
#print(cm_test)
```

```python
cr_train = classification_report(y_sm, y_train_pred)
print("TRAIN:", cr_train)

fig, ax=plt.subplots(1,2, figsize=(14,4))
plot_confusion_matrix(lsvc,X_train_mod,y_sm, ax=ax[0], values_format = 'd')
ax[0].set_title('Train')

plot_confusion_matrix(lsvc, X_test_mod,y_test, ax=ax[1], values_format = 'd')
ax[1].set_title('Test')

cr_test = classification_report(y_test, y_test_pred)
print("TEST:", cr_test)

plt.show()
```

```python
cv_scores_train = cross_val_score(lsvc, X_train_mod, y_sm, cv=5)
print("Cross Validation score: ", cv_scores_train)
print("Score stats: %0.2f accuracy with a standard deviation of %0.2f" % (cv_scores_train.mean(), cv_scores_train.std()))


#cv_scores_test = cross_val_score(lsvc, X_test_mod, y_test, cv=5)
#print("Cross Validation score: ", cv_scores_test)
#print("Score stats: %0.2f accuracy with a standard deviation of %0.2f" % (cv_scores_test.mean(), cv_scores_test.std()))
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)

plt.grid()

plt.plot(fpr, tpr, label=" AUC TEST ="+str(auc(fpr, tpr)))
plt.plot([0,1],[0,1],'g--')
plt.legend()
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("AUC(ROC curve)")
plt.grid(color='black', linestyle='-', linewidth=0.5)
plt.show()
```
