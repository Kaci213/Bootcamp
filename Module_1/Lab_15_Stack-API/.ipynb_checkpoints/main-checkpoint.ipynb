```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1669852861, todate=1672531199, sort='votes')
questions
```

```python
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate=1669852861, todate=1672531199, sort='votes')
answers
```
